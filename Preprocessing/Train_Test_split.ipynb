```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
```

```python
dataset_sentiment = pd.read_json('full_data\\sentiment_score.json', lines=True)
dataset_sentiment = dataset_sentiment.rename(columns={'sentiment_score': 'roberta_score'})
dataset_sentiment
```

```python
#take data from Starbucks, McDonald's and Dunkin' to be our BI data for the final comparison
BI_df = dataset_sentiment[dataset_sentiment["name"].str.contains("Starbucks", na=False)]
#BI_df.drop(columns=['sentiment', 'roberta_score'], inplace=True)
BI_df
```

```python
#take rest data of the data to be our training data for the models
train_model_df = dataset_sentiment[~dataset_sentiment["name"].str.contains("Starbucks", na=False)]
train_model_df
```

```python
df = train_model_df[['business_id','text', 'roberta_score']]
df = df.dropna(subset=['business_id','text', 'roberta_score'])
df
```

```python
#Split into Train and Test (80-20)
train_df, test_df = train_test_split(
    df, 
    test_size=0.2, 
    random_state=42, 
    stratify=df['roberta_score'])

print(f"Training data shape: {train_df.shape}")
print(f"Test data shape: {test_df.shape}")
```

```python
#Split into Train and Validation (80-20)
X_train, X_val, y_train, y_val = train_test_split(
    train_df['text'].values, 
    train_df['roberta_score'].values, 
    test_size=0.2, 
    random_state=42, 
    stratify=train_df['roberta_score']
)

# Prepare the test set
X_test = test_df['text'].values
y_test = test_df['roberta_score'].values

print(f"Training samples: {len(X_train)}, Validation samples: {len(X_val)}, Test samples: {len(X_test)}")
```
