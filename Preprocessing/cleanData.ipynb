```python
#Review.json  - Properties taken : user_id, business_id, stars, useful, funny, cool, text
reviewCols = ['user_id', 'business_id', 'stars', 'useful', 'funny', 'cool', 'text', 'date']

with open('D:\\CBS_MSc_Docs\\Courses\\Sem3\\AIML\\data\\yelp_academic_dataset_review.json', encoding='utf-8') as f:
    for line in f:
        record = json.loads(line)
        # Extract only the desired properties
        data.append({col: record.get(col) for col in reviewCols})

df_review = pd.DataFrame(data)

f.close()
data = []
```

```python
#Read Business.json

businessCols = ['business_id','name','review_count','categories','address', 'city', 'state', 'postal_code']

with open('D:\\CBS_MSc_Docs\\Courses\\Sem3\\AIML\\data\\yelp_academic_dataset_business.json', encoding='utf-8') as f:
    for line in f:
        record = json.loads(line)
        # Extract only the desired properties
        data.append({col: record.get(col) for col in businessCols})

# Flatten the JSON
#df_business = pd.json_normalize(data)

#Remove columns not required
df_business = pd.DataFrame(data)

f.close()
data = []

#df_business.columns
```
